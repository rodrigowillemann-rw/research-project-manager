# Research Project Manager - Portfolio de Editais

Exemplo de analise para priorizar projetos de pesquisa e inovacao.

```python
import pandas as pd
df = pd.read_csv('../data/editais_projetos_exemplo.csv')
df['urgencia'] = df['prazo_submissao_dias'].apply(lambda x: 3 if x <= 20 and x > 0 else 2 if x <= 45 and x > 0 else 1)
df['score_priorizacao'] = df['valor_solicitado_kbrl'] * 0.01 + df['parceiros'] * 5 + df['urgencia'] * 10
df.sort_values('score_priorizacao', ascending=False)[['projeto','area','status','score_priorizacao']]
```
